```python
SELECT column_name, data_type
FROM INFORMATION_SCHEMA.columns
WHERE table_name = 'economies'
```

```python
SELECT 
    year,
    income_group,
    COUNT(DISTINCT code) AS country_count
FROM 
    world.economies
GROUP BY 
    year, income_group
ORDER BY 
    year, income_group;
```

```python
WITH GDPGrowth AS (
    SELECT 
        econ_id,
        code,
        year,
        gdp_percapita,
        LAG(gdp_percapita) OVER (PARTITION BY code ORDER BY year) AS prev_gdp_percapita
    FROM 
        world.economies
)
SELECT 
    econ_id,
    code,
    year,
    gdp_percapita,
    (gdp_percapita - prev_gdp_percapita) / prev_gdp_percapita * 100 AS gdp_growth_percentage
FROM 
    GDPGrowth
WHERE 
    prev_gdp_percapita IS NOT NULL
ORDER BY 
    gdp_growth_percentage DESC
LIMIT 10;
```

```python
SELECT code, year, gdp_percapita
FROM world.economies
ORDER BY world.economies.code ASC
```

```python
SELECT *
FROM world.economies
WHERE world.economies.income_group != 'High income'
  AND world.economies.gdp_percapita IS NOT NULL
  AND world.economies.year != 2010 
  AND world.economies.unemployment_rate IS NOT NULL
ORDER BY gdp_percapita DESC
```
